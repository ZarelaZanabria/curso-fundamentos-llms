```python
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
tokens = tokenizer.tokenize("El gato maulla, entonces el perro que hace ?")
```

```python
tokens
```

```python
import torch
from transformers import BertModel

model = BertModel.from_pretrained("bert-base-uncased")
```

```python
king_token_id = tokenizer.convert_tokens_to_ids(["conejo"])[0]
king_embedding = model.embeddings.word_embeddings(torch.tensor([king_token_id]))

man_token_id = tokenizer.convert_tokens_to_ids(["rat√≥n"])[0]
man_embedding = model.embeddings.word_embeddings(torch.tensor([king_token_id]))

woman_token_id = tokenizer.convert_tokens_to_ids(["muerde"])[0]
woman_embedding = model.embeddings.word_embeddings(torch.tensor([king_token_id]))

queen_token_id = tokenizer.convert_tokens_to_ids(["pez"])[0]
queen_embedding = model.embeddings.word_embeddings(torch.tensor([queen_token_id]))
```

```python
cos = torch.nn.CosineSimilarity()
```

```python
similarity = cos(king_embedding, queen_embedding)
print(similarity[0])
```

# Task
Find the animals most similar to "cat" using BERT embeddings and cosine similarity.

## Obtener el embedding de "gato"

### Subtask:
Utilizar el tokenizador y el modelo BERT cargados previamente para obtener el embedding de la palabra "gato".


**Reasoning**:
Convert the word "gato" to its token ID and get its embedding using the pre-trained BERT tokenizer and model.



```python
cat_token_id = tokenizer.convert_tokens_to_ids(["gato"])[0]
cat_embedding = model.embeddings.word_embeddings(torch.tensor([cat_token_id]))
```

## Preparar una lista de palabras para comparar

### Subtask:
Definir una lista de palabras de animales para comparar con "gato".


**Reasoning**:
Define a list of animal words to compare with the "cat" embedding.



```python
animal_words = ["perro", "conejo", "rat√≥n", "p√°jaro", "pez", "caballo", "vaca", "cerdo", "oveja", "tigre"]
```

## Obtener embeddings para las palabras de comparaci√≥n

### Subtask:
Obtener los embeddings para cada palabra en la lista de animales.


**Reasoning**:
Iterate through the animal_words list, get the token ID and embedding for each word, and store them in a dictionary.



```python
animal_embeddings = {}
for animal in animal_words:
  token_id = tokenizer.convert_tokens_to_ids([animal])[0]
  embedding = model.embeddings.word_embeddings(torch.tensor([token_id]))
  animal_embeddings[animal] = embedding

display(animal_embeddings)
```

## Calcular la similitud

### Subtask:
Calcular la similitud del coseno entre el embedding de "gato" y los embeddings de cada palabra en la lista de animales.


**Reasoning**:
Initialize an empty dictionary to store similarity scores, iterate through the animal embeddings, calculate the cosine similarity between the cat embedding and each animal embedding, store the results in the dictionary, and finally display the dictionary.



```python
similarity_scores = {}
for animal, embedding in animal_embeddings.items():
  similarity = cos(cat_embedding, embedding)
  similarity_scores[animal] = similarity.item()

display(similarity_scores)
```

## Mostrar los resultados

### Subtask:
Ordenar las palabras por similitud y mostrar los animales m√°s parecidos a "gato".


**Reasoning**:
Sort the similarity scores and print the results to show the most similar animals.



```python
sorted_similarity = sorted(similarity_scores.items(), key=lambda item: item[1], reverse=True)

print("Animales m√°s similares a 'gato':")
for animal, score in sorted_similarity:
    print(f"- {animal}: {score:.4f}")
```

## Summary:

### Data Analysis Key Findings

*   The BERT embeddings were successfully obtained for the word "gato" and a list of other animal words.
*   Cosine similarity was calculated between the embedding of "gato" and the embeddings of each animal word.
*   Unexpectedly, all animal words in the list ("perro", "conejo", "rat√≥n", "p√°jaro", "pez", "caballo", "vaca", "cerdo", "oveja", "tigre") showed a perfect cosine similarity score of 1.0000 with "gato".

### Insights or Next Steps

*   The perfect similarity score of 1.0000 across all animals suggests a potential issue with the BERT embeddings obtained or the method used to generate them for individual words in this specific context. It would be beneficial to investigate why these embeddings are resulting in identical vectors for different animal names.
*   Consider using sentence-level embeddings or contextualized embeddings if the goal is to capture more nuanced semantic relationships between words, as single-word embeddings might not always differentiate concepts effectively depending on the model and task.


## Utilizar palabras en ingl√©s para obtener embeddings m√°s significativos

**Reasoning**:
The previously used Spanish words were tokenized to the same unknown token ID, leading to identical embeddings and a cosine similarity of 1.0 for all comparisons. Since the `bert-base-uncased` model is primarily trained on English, using the English translations of the animal words will result in distinct token IDs and embeddings, allowing for a more accurate representation of semantic similarity.

```python
# Define the animal words in English
animal_words_en = ["dog", "rabbit", "mouse", "bird", "fish", "horse", "cow", "pig", "sheep", "tiger"]

# Get the token ID and embedding for "cat" in English
cat_token_id_en = tokenizer.convert_tokens_to_ids(["cat"])[0]
cat_embedding_en = model.embeddings.word_embeddings(torch.tensor([cat_token_id_en]))

# Get embeddings for the English animal words
animal_embeddings_en = {}
for animal in animal_words_en:
  token_id_en = tokenizer.convert_tokens_to_ids([animal])[0]
  embedding_en = model.embeddings.word_embeddings(torch.tensor([token_id_en]))
  animal_embeddings_en[animal] = embedding_en

# Calculate cosine similarity between "cat" and other animals in English
similarity_scores_en = {}
for animal, embedding in animal_embeddings_en.items():
  similarity_en = cos(cat_embedding_en, embedding)
  similarity_scores_en[animal] = similarity_en.item()

# Sort and display the results
sorted_similarity_en = sorted(similarity_scores_en.items(), key=lambda item: item[1], reverse=True)

print("Animales m√°s similares a 'cat' (en ingl√©s):")
for animal, score in sorted_similarity_en:
    print(f"- {animal}: {score:.4f}")
```
